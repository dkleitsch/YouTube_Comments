```python
import pandas as pd

# Create filepaths for importing data
file_pathftm = "Comments/ftm_comments.csv"
file_pathmtf = "Comments/mtf_comments.csv"
file_pathnon = "Comments/nonbinary_comments.csv"
```

```python
# Read ftm, mtf, nonbinary into dataframes
ftm_df = pd.read_csv(file_pathftm)
ftm_df.head()
```

```python
mtf_df = pd.read_csv(file_pathmtf)
mtf_df.head()
```

```python
nonbinary_df = pd.read_csv(file_pathnon)
nonbinary_df.head()
```

```python
# Concatenate ftm and mtf
trans_df = pd.concat([ftm_df, mtf_df], axis=0)
trans_df
```

```python
# Concatenate trans and nonbinary
transgender_df = pd.concat([trans_df, nonbinary_df], axis=0)
transgender_df
```

```python
ftm_df.to_csv("ftm_comments.csv", index=False)
```
