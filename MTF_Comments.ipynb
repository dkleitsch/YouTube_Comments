```python
import pandas as pd

# Create filepaths for importing data
file_pathgg = "Comments/gigi_final.csv"
file_pathgg2 = "Comments/gigi2_final.csv"
file_pathm = "Comments/matilda_final.csv"
file_pathm2 = "Comments/matilda2_final.csv"
```

```python
# Read gigi, gigi2, matilda, and matilda2 into dataframes
gg_df = pd.read_csv(file_pathgg)
gg_df.head()
```

```python
gg2_df = pd.read_csv(file_pathgg2)
gg2_df.head()
```

```python
m_df = pd.read_csv(file_pathm)
m_df.head()
```

```python
m2_df = pd.read_csv(file_pathm2)
m2_df.head()
```

```python
# Drop unnamed column
gg_df.drop(gg_df.filter(regex="Unname"), axis=1, inplace=True)
gg_df
```

```python
gg2_df.drop(gg2_df.filter(regex="Unname"), axis=1, inplace=True)
gg2_df
```

```python
m_df.drop(m_df.filter(regex="Unname"), axis=1, inplace=True)
m_df.head()
```

```python
m2_df.drop(m2_df.filter(regex="Unname"), axis=1, inplace=True)
m2_df.head()
```

```python
# Concatenate gg and gg2 
ggs_df = pd.concat([gg_df, gg2_df], axis=0)
ggs_df
```

```python
# Concatenate m and m2
mat_df = pd.concat([m_df, m2_df], axis=0)
mat_df
```

```python
# Concatenate ggs and mat
mtf_df = pd.concat([ggs_df, mat_df], axis=0)
mtf_df
```

```python
# Drop viewer rating column
mtf_df.drop(columns="Viewer Rating", inplace=True)
mtf_df
```

```python
mtf_df.drop(mtf_df.filter(regex="Unname"), axis=1, inplace=True)
mtf_df
```

```python
# Add new column with host gender identity (2=MaletoFemale)
mtf_df['Host Gender ID']=2
mtf_df
```

```python
mtf_df.to_csv("mtf_comments.csv", index=False)
```
