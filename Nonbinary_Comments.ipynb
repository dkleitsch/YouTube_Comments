```python
import pandas as pd

# Create filepaths for importing data
file_path = "Comments/riley_final.csv"
file_path2 = "Comments/riley2_final.csv"
```

```python
# Read riley and riley2 into dataframes
riley_df = pd.read_csv(file_path)
riley_df.head()
```

```python
riley2_df = pd.read_csv(file_path2)
riley2_df.head()
```

```python
# Concatenate gg and gg2 
nonbinary_df = pd.concat([riley_df, riley2_df], axis=0)
nonbinary_df
```

```python
# Drop viewer rating column
nonbinary_df.drop(columns="Viewer Rating", inplace=True)
nonbinary_df
```

```python
nonbinary_df.drop(nonbinary_df.filter(regex="Unname"), axis=1, inplace=True)
nonbinary_df
```

```python
# Add new column with host gender identity (3=Nonbinary)
nonbinary_df['Host Gender ID']=3
nonbinary_df
```

```python
nonbinary_df.to_csv("nonbinary_comments.csv", index=False)
```
